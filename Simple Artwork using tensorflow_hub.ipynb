```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import tensorflow_hub as tf_hub
```

```python
def load_image(img_path):
    img = tf.io.read_file(img_path)
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = img[tf.newaxis, :]
    return img
```

```python
from google.colab import drive
drive.mount('/content/drive/')
```

```python
content_image = load_image("/content/drive/MyDrive/01-How To Create Artwork Using Neural Networks In Minutes/Content.png")
style_image = load_image("/content/drive/MyDrive/01-How To Create Artwork Using Neural Networks In Minutes/Style.png")
```

```python
fig = plt.figure(figsize=(10,6))

fig.add_subplot(1,2, 1)
plt.imshow(np.squeeze(content_image))

fig.add_subplot(1,2, 2)
plt.imshow(np.squeeze(style_image))
plt.show(block=True)
```

```python
model = tf_hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

```python
stylized_image = model(tf.constant(content_image), tf.constant(style_image))[0]
```

```python
f = plt.figure(figsize=(10,6))
plt.imshow(np.squeeze(stylized_image))
plt.title('Generated Artwork')
plt.show(block=True)
```
